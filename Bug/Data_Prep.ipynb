```python
import six
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.metrics import confusion_matrix#,plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.preprocessing import MinMaxScaler

Inventory=pd.read_csv('train_inventory.csv')
```

```python
Inventory.describe()
```

```python
Inventory.isna().sum()
```

```python
Inventory['lead_time'].fillna(Inventory['lead_time'].mean(),inplace=True)
```

```python
Inventory['went_on_backorder'].value_counts()
```

```python
x=Inventory.drop(['went_on_backorder','sku'],axis=1)
y=Inventory['went_on_backorder']
x.head()
y.head()
x=pd.get_dummies(x)
y=pd.get_dummies(y)
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.3,random_state=40)
X_train.reset_index(inplace=True)
X_test.reset_index(inplace=True)
X_train.drop(['index'],axis=1,inplace=True)
X_test.drop(['index'],axis=1,inplace=True)
Y_train=Y_train.reset_index()
Y_test=Y_test.reset_index()
Y_train.drop(['index','No'],axis=1,inplace=True)
Y_test.drop(['index','No'],axis=1,inplace=True)
```

```python
X_train.head()
```

```python
scaler=MinMaxScaler()
X_train_scaled=pd.DataFrame(scaler.fit_transform(X_train),columns=X_train.columns)
X_test_scaled=pd.DataFrame(scaler.fit_transform(X_test),columns=X_train.columns)
model=LogisticRegression(max_iter=1000)
model.fit(X_train_scaled,Y_train)
Prediction=model.predict(X_test_scaled)

```

```python
X_train_scaled
```

```python
cm=confusion_matrix(Y_test,Prediction)
print(cm)
```

```python
print(classification_report(Y_test,Prediction))
```

```python
def score(model1,request):
    from category_encoders import OneHotEncoder
    request_type = "payload"
    import numpy
    import pandas as pd
    payload = request.json
    
    if payload['potential_issue'] == 'Yes':
        payload['potential_issue_No'] = 0   
        payload['potential_issue_Yes'] = 1       
    else:
        payload['potential_issue_No'] = 1   
        payload['potential_issue_Yes'] = 0

    if payload['deck_risk'] == 'Yes':
       payload['deck_risk_No'] = 0
       payload['deck_risk_Yes'] = 1
      
    else:
        payload['deck_risk_No'] = 1   
        payload['deck_risk_Yes'] = 0
   

    if payload['oe_constraint'] == 'Yes':
       payload['oe_constraint_No'] = 0
       payload['oe_constraint_Yes'] = 1
       
    else:
       payload['oe_constraint_No'] = 1
       payload['oe_constraint_Yes'] = 0
       

    if payload['ppap_risk'] == 'Yes':
       payload['ppap_risk_No'] = 0
       payload['ppap_risk_Yes'] = 1
       
    else:
       payload['ppap_risk_No'] = 1 
       payload['ppap_risk_Yes'] = 0
        

    if payload['stop_auto_buy'] == 'Yes':
       payload['stop_auto_buy_No'] = 0 
       payload['stop_auto_buy_Yes'] = 1
       
    else:
       payload['stop_auto_buy_No'] = 1 
       payload['stop_auto_buy_Yes'] = 0
        

    if payload['rev_stop'] == 'Yes':
       payload['rev_stop_No'] = 0 
       payload['rev_stop_Yes'] = 1
       
    else:
       payload['rev_stop_No'] = 1 
       payload['rev_stop_Yes'] = 0
                

    payload.pop('potential_issue')
    payload.pop('deck_risk')
    payload.pop('oe_constraint')
    payload.pop('ppap_risk')
    payload.pop('stop_auto_buy')
    payload.pop('rev_stop')

    print(payload)
        
    dataframe=pd.DataFrame(payload,index=range(0,1))
#    dataframe1=pd.get_dummies(dataframe)
    pred=model1.predict(dataframe)
    return pred.tolist()
```
